```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
URL = 'http://quotes.toscrape.com/'
```

```python
page = requests.get(URL)
page.status_code
```

```python
htmlcode = page.text
soup = BeautifulSoup(htmlcode)

htmlcode
```

```python
content = soup.find('div',attrs={'class' : 'container'})
text=content.text
print(text)

r1 = text.split("\n")
#print(r1)

r2 = []
for i in r1:
  if i!='' and i!='\r':
    r2.append(i)
#print(r2)

r3 = []
for i in r2:
  k = i.strip()
  r3.append(k)
#print(r3)

r4 = []
for i in r3:
  if i!='':
    r4.append(i)
#print(r4)

data=[]
for i in r4:
    if len(i)>10:
     data.append(i)

quotes = []
authors = []
for i in range(len(data)):
    if data[i].startswith("“") and (i + 1) < len(data) and data[i + 1].startswith("by"):
        quotes.append(data[i])
        authors.append(data[i + 1].replace("by ", ""))

print(quotes)
print(authors)
```

```python
df = pd.DataFrame({"Quote": quotes, "Author": authors})
df
```

```python
df.to_csv('products.csv', header=True, index=False)
df
```
